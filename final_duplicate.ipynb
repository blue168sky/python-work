```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('final.csv')
print(df)
```

```python
for i in range(40):
    sample = df.sample(n=1, replace=True, axis=0)
    df = pd.concat([df,sample], axis=0, ignore_index=True)
print(df)
    
```

```python
x_train = df.drop(['DATE','RESULT'],axis = 1)
y_train = df['RESULT']
print(x_train)
print(y_train)
```

```python
from sklearn.decomposition import PCA 
pca=PCA(n_components=4)
```

```python
pca.fit(x_train)
x_train_pca = pca.transform(x_train)
print("original shape:   ", x_train.shape)
print("transformed shape:", x_train_pca.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
model = LogisticRegression()
model.fit(x_train_pca, y_train)
print(model.coef_)
print(model.intercept_)
```

```python
df1 = pd.read_csv('final_test.csv')
df_test = df1.drop(df.index[0:5])
x_test = df_test.drop(['DATE','RESULT'],axis = 1)
print(x_test)

pca_test=PCA(n_components=4)
pca_test.fit(x_test)
x_test_pca = pca_test.transform(x_test)

print("original shape:   ", x_test.shape)
print("transformed shape:", x_test_pca.shape)
```

```python
print(x_test_pca)
```

```python
prob =np.round(model.predict_proba(x_test_pca),3)
print(np.delete(prob,1,1))
```
