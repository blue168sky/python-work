# Boston Dataset


```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LinearRegression
df = datasets.load_boston()
df1 = pd.DataFrame(df['data'],columns=df['feature_names'])
df2 = pd.DataFrame(df['target'],columns=['DESCR'])
Boston = pd.concat([df1,df2],axis = 1)
print(Boston)
```

地區犯罪率與當地房價關係

```python
x = Boston[['CRIM','LSTAT']]
y = Boston['DESCR']
lm = LinearRegression()
model=lm.fit(x, y)
print('迴歸係數',lm.intercept_, lm.coef_[0], lm.coef_[1])
```

```python
pd.DataFrame(lm.predict(x), columns = ['prediction'])
```
